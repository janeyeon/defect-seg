```python
root = "/home2/hyun/DefectSeg/data/Vision_v1"
domains = os.listdir(root)
print(domains)
```

```python
n_data_total = 0
for domain in domains:
    if domain is not "Cable":
        continue
    if os.path.isdir(f"{root}/{domain}"):

        data_types = os.listdir(f"{root}/{domain}/{domain}")
        print(domain)
        n_data = 0
        
        for data_type in data_types:
            data_fns = os.listdir(f"{root}/{domain}/{domain}/{data_type}")
            data_fns = [fn for fn in data_fns if ("._" not in fn) and (".json" not in fn)]
            n_data += len(data_fns)
            # print(len(data_fns))
            
            for data_fn in data_fns[:20]:
                img = cv2.imread(f"{root}/{domain}/{domain}/{data_type}/{data_fn}")
                print(img.shape)
        
        print(n_data)
        print()
    n_data_total += n_data
print("Total")
print(n_data_total)
```
